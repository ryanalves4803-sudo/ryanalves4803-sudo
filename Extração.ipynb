### Entendendo os Dados

Ao analisar esse dataset, dá para perceber que ele reúne informações sobre programas de pós-graduação no Brasil, provavelmente ligados à CAPES. Pelas colunas apresentadas, é possível entender vários aspectos importantes sobre as universidades e seus programas.

Primeiro, aparecem dados da instituição e do programa, mostrando o nome da universidade e qual curso de pós-graduação ela oferece. Também existe a coluna de nível, que indica se o programa é de mestrado, doutorado ou outro tipo. Já a coluna “tem doutorado” mostra se aquela instituição oferece doutorado além do mestrado.

Outro ponto interessante são as informações sobre os docentes, ou seja, os professores do programa. O dataset parece separar os docentes permanentes e colaboradores, permitindo entender melhor como é formada a equipe acadêmica.
Além disso, existem várias colunas relacionadas à produção acadêmica, como artigos e livros classificados em categorias A1, A2, B1 e outras. Essas classificações provavelmente seguem o sistema Qualis, utilizado para avaliar a qualidade das produções científicas.

Também há dados sobre os estudantes, como quantidade de bolsas, número de alunos no mestrado e no doutorado. Algumas colunas com nomes começando em “per_comaluno_” parecem indicar a porcentagem de trabalhos feitos em parceria entre professores e alunos, mostrando o nível de participação dos estudantes nas pesquisas.

Os próximos passos da análise seriam começar pela limpeza dos dados, verificando se existem valores vazios, informações repetidas ou inconsistentes. Isso é importante para garantir que os resultados da análise sejam confiáveis.

```python
print('Valores nulos por coluna antes do tratamento:')
print(dados_novos.isnull().sum()[dados_novos.isnull().sum() > 0])
```

### Tratamento de Valores Nulos

Ao analisar os valores nulos do dataset, foi possível identificar que algumas colunas apresentam dados faltando. A coluna Dissertações possui 4 valores nulos, enquanto as colunas per_comaluno_A1 até per_comaluno_NA também apresentam 4 valores ausentes cada uma. Já a coluna Teses chama mais atenção, pois possui 50 valores nulos, o que representa uma quantidade muito alta em relação ao total de registros.

Para tratar esses dados faltantes, algumas estratégias podem ser utilizadas. No caso das colunas Dissertações e per_comaluno_, como a quantidade de valores nulos é pequena, uma boa alternativa é preencher esses espaços utilizando a mediana ou a média dos valores da própria coluna. Isso ajuda a manter os dados mais equilibrados sem causar grandes alterações na análise. Outra possibilidade seria preencher com zero, principalmente se a ausência do valor significar ausência de produção acadêmica.

Já a coluna Teses exige mais cuidado, porque possui muitos dados faltantes. Preencher tudo com média ou mediana poderia distorcer bastante os resultados. Nesse caso, a ausência de valores pode indicar que o programa não possui doutorado ou que não houve produção de teses naquele período analisado. Por isso, preencher com zero parece ser uma solução mais coerente, já que provavelmente representa a inexistência de teses. Outra alternativa seria remover a coluna da análise, caso ela não seja tão importante para os objetivos do estudo.

Como primeira etapa do tratamento, a decisão foi preencher as colunas Dissertações e per_comaluno_ (A1 até NA) utilizando a mediana. Já a coluna Teses será preenchida com zero, considerando que a falta de informação provavelmente representa ausência de produção de teses.

```python
# Preenchendo valores nulos para 'Dissertacoes' com a mediana
dados_novos['Dissertacoes'] = dados_novos['Dissertacoes'].fillna(dados_novos['Dissertacoes'].median())

# Preenchendo valores nulos para 'Teses' com 0, considerando que a ausência pode significar 0 teses
dados_novos['Teses'] = dados_novos['Teses'].fillna(0)

# Preenchendo valores nulos para as colunas 'per_comaluno_' com a mediana
per_comaluno_cols = [col for col in dados_novos.columns if 'per_comaluno_' in col]
for col in per_comaluno_cols:
    dados_novos[col] = dados_novos[col].fillna(dados_novos[col].median())

print('\nValores nulos por coluna após o tratamento:')
print(dados_novos.isnull().sum()[dados_novos.isnull().sum() > 0])
```

### Programas com Maior Número de Artigos em Conferência

```python
artigos_por_programa = dados_novos.groupby('Programa')['Artigos em conf'].sum().reset_index()
programas_top_artigos = artigos_por_programa.sort_values(by='Artigos em conf', ascending=False)

print('Programas com o maior número de artigos em conferência:')
display(programas_top_artigos.head(10))
```

### Análise Exploratória e Visualização

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações para os gráficos
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

# Distribuição dos Níveis dos Programas
sns.countplot(x='Nível', data=dados_novos, hue='Nível', palette='viridis', legend=False)
plt.title('Distribuição dos Níveis dos Programas de Pós-Graduação')
plt.xlabel('Nível do Programa')
plt.ylabel('Número de Programas')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Alunos Colaboradores
plt.subplot(1, 2, 1)
sns.histplot(dados_novos['Docentes colaboradores'], kde=True, color='skyblue')
plt.title('Distribuição de Docentes Colaboradores')
plt.xlabel('Número de Docentes Colaboradores')
plt.ylabel('Frequência')

# Alunos Permanentes
plt.subplot(1, 2, 2)
sns.histplot(dados_novos['Docentes permanentes'], kde=True, color='lightcoral')
plt.title('Distribuição de Docentes Permanentes')
plt.xlabel('Número de Docentes Permanentes')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```

Esses gráficos nos dão uma ideia inicial da composição dos programas. O próximo passo pode ser explorar as correlações entre as variáveis numéricas, como docentes e produção, ou a produção por nível de programa.

### Análise de Correlação

Verificamos que esse mapa de calor mostra a correlação entre as variáveis numéricas. Valores próximos de 1 indicam uma forte correlação positiva, enquanto valores próximos de -1 indicam uma forte correlação negativa. Valores próximos de 0 indicam pouca ou nenhuma correlação linear.

### Análise de Correlação

```python
# Código pra selecionar apenas colunas numéricas para a correlação
numeric_cols = dados_novos.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = dados_novos[numeric_cols].corr()

plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Matriz de Correlação das Variáveis Numéricas')
plt.show()
```

Este mapa de calor mostra a correlação entre as variáveis numéricas. Valores próximos de 1 indicam uma forte correlação positiva, enquanto valores próximos de -1 indicam uma forte correlação negativa. Valores próximos de 0 indicam pouca ou nenhuma correlação linear.
